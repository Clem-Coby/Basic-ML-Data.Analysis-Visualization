```python
#Standard imports
import numpy as np
import pandas as pd
```

```python
#For Stats
from scipy import stats
```

```python
#For Plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Magic Functions to diplay graphs inline in the notebook
%matplotlib inline
```

```python
dataset1 = np.random.randn(80)

plt.hist(dataset1)

plt.show()
```

```python
dataset2 = np.random.randn(100)

plt.hist(dataset2, color="indianred")

plt.show()
```

```python
# Both Histograms on the same plot
# To maintain transprency, use alpha. 
# Since both have different types of datapoints 80 and 100, we've to normalize the plots but setting normed to True
# Default bins are 10. We can explicitly mention bins required.

plt.hist(dataset1,normed=True, alpha=0.5, color="indianred", bins=20)
plt.hist(dataset2, normed=True, alpha=0.5, bins=20)

plt.show()
```

```python
#Plotting Two Data Points Using Seaborn

data1 = np.random.rand(1000)
data2 = np.random.rand(1000)
```

```python
sns.jointplot(data1,data2)
plt.show()
```

```python
sns.jointplot(data1,data2, kind='hex')
plt.show()
```
